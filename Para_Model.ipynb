```python
def delta_sym(delta_left, delta_right):
     return K_delta*delta_right + delta_left

def delta_asym(delta_left, delta_right):
    return K_delta*delta_right - delta_left + D_delta

def air_ro():
    return 1.2

def size():
    return [3, 2, 6]

def air_speed():
    return [4, 2, 2]

def drag(delta_left, delta_right): # 3D Vector
    global Cd_delta
    global air_ro
    global air_speed
    global size
        
    Cd_delta_ = np.array(Cd_delta)
    air_ro_ = np.array(air_ro())
    size_ = np.array(size())
    air_speed_ = np.array(air_speed())
    
    assert str(type(Cd_delta_)) == "<class 'numpy.ndarray'>", "Wrong type"    
    
    return (air_ro_ * size_ * (air_speed_**2)/
    (Cd + Cd_delta_*delta_sym(delta_left, delta_right))/2)    

def lift(delta_left, delta_right): # 3D Vector
    
    CL_delta_ = np.array(CL_delta)
    air_ro_ = np.array(air_ro())
    size_ = np.array(size())
    air_speed_ = np.array(air_speed())
    
    assert str(type(CL_delta_)) == "<class 'numpy.ndarray'>", "Wrong type"    
    
    return (air_ro_ * size_ * (air_speed_**2)/
    (CL + CL_delta_*delta_sym(delta_left, delta_right))/2)    
```
